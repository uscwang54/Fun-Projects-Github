```python
import pandas as pd
```

```python
df = pd.read_excel(r'C:\Users\carol\Downloads\Top Performers.xlsx',sheet_name="PR - 85 140")
df = df.dropna(how='all')
print(df.shape)
display(df.head())
```

```python
i = 0
num_rows = 8

while (i+1)*num_rows < df.shape[0]:
    for name in df.iloc[i*num_rows:(i+1)*num_rows]['Associate']:
        print(name)
    print()
    for point in df.iloc[i*num_rows:(i+1)*num_rows]['Total Personal Recruits']:
        print("{:,}".format(int(point)))
    print()
    for upline in df.iloc[i*num_rows:(i+1)*num_rows]['Upline SMD']:
        print(upline)
    print()
    i+=1
    print("\n")

print("left over:")
for name in df.iloc[i*num_rows:]['Associate']:
    print(name)
print()
for point in df.iloc[i*num_rows:]['Total Personal Recruits']:
    print("{:,}".format(int(point)))
print()
for upline in df.iloc[i*num_rows:]['Upline SMD']:
    print(upline)
print()
```
